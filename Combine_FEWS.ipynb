```python
import os
import numpy as np
import pandas as pd
import geopandas as gpd
from functools import reduce
```

```python
CS_Oct22 = gpd.read_file('BF_202210_CS.shp')[['CS', 'geometry']]
CS_Oct22['FID'] = range(0,len(CS_Oct22))
CS_Oct22
```

```python
DFlist = [pd.DataFrame(gpd.read_file('BF_202206_CS.shp'))[['CS']],
          pd.DataFrame(gpd.read_file('BF_202202_CS.shp'))[['CS']],
          pd.DataFrame(gpd.read_file('BF_202110_CS.shp'))[['CS']],
          pd.DataFrame(gpd.read_file('BF_202106_CS.shp'))[['CS']]]
for item in DFlist:
    print(item.head())
    item['FID'] = range(0,len(item))
    print(item.head())
```

```python
DFlist[0] = DFlist[0].rename(columns={"CS": "01Jun22"})
DFlist[1] = DFlist[1].rename(columns={"CS": "01Feb22"})
DFlist[2] = DFlist[2].rename(columns={"CS": "01Oct21"})
DFlist[3] = DFlist[3].rename(columns={"CS": "01Jun21"})
for item in DFlist:
    print(item.head())
```

```python
CS_Oct22 = CS_Oct22.rename(columns={"CS": "01Oct22"})
print(CS_Oct22.head())
```

```python
DFlist = [CS_Oct22] + DFlist

FEWS = reduce(lambda left,right: pd.merge(left,right,on=['FID'], how='left'), DFlist)
print(FEWS.sample(5), FEWS.info())
```

```python
FEWS = FEWS.melt(id_vars=['FID', 'geometry'], var_name='time_start', value_name='CS')
print(FEWS.sample(5), FEWS.info())
```

```python
FEWS['time_end'] = 'placeholder'

FEWS.loc[FEWS['time_start'] == "01Jun21", 'time_end'] = '30Sep21'
FEWS.loc[FEWS['time_start'] == "01Oct21", 'time_end'] = '31Jan22'
FEWS.loc[FEWS['time_start'] == "01Feb22", 'time_end'] = '31May22'
FEWS.loc[FEWS['time_start'] == "01Jun22", 'time_end'] = '30Sep22'
FEWS.loc[FEWS['time_start'] == "01Oct22", 'time_end'] = '31Jan23'
```

```python
FEWS.to_file('FEWS_CurrentSit_2021to2022.shp')
```
